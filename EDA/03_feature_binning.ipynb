### 2. Basic Processing

```python
df = pd.read_csv('data/processed/ChurnModelling_Missing_Values_Handled.csv')
df.head()
```

350 - 580 => 'Poor'
580 - 670 => 'Fair'
670 - 740 => 'Good' 
740 - 800 => 'Very Good'
800 - 850 => Excellent

```python
def custom_bining_credit_score(score):
    if score < 580:
        return 'Poor'
    if score < 670:
        return 'Fair'
    if score < 740:
        return 'Good'
    if score < 800:
        return 'Very Good'
    if score <= 850:
        return 'Excellent'
    else:
        assert True, "Credit Score can't go beyond 850"
```

```python
df['CreditScoreBins'] = df['CreditScore'].apply(custom_bining_credit_score)
df.head()
```

```python
df.to_csv('data/processed/ChurnModelling_Binning_Applied.csv', index=False)
```
